# 03 – Test

This notebook connects to the deployed Azure AI Foundry model and validates end-to-end
feasibility by running a series of chat prompts using the **`azure-ai-inference` SDK**.

**Pre-requisite:** run `02_configure.ipynb` first.

---

## Step 1 – Load environment variables

```python
import os
from dotenv import load_dotenv

load_dotenv(dotenv_path="../env/.env")

endpoint   = os.environ["AZURE_AI_ENDPOINT"]
api_key    = os.environ["AZURE_AI_API_KEY"]
deployment = os.environ.get("AZURE_AI_DEPLOYMENT", "chat")

print(f"Endpoint   : {endpoint}")
print(f"Deployment : {deployment}")
```

## Step 2 – Create the Azure AI Inference client

Using the new Azure AI Foundry `ChatCompletionsClient` from `azure-ai-inference`.

```python
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(api_key),
)
print("✅  Client created")
```

## Step 3 – Helper function

A small wrapper to keep prompt calls readable throughout the notebook.

```python
def chat(user_message: str, system_message: str = "You are a helpful assistant.") -> str:
    """Send a single-turn chat prompt and return the response text."""
    response = client.complete(
        model=deployment,
        messages=[
            SystemMessage(content=system_message),
            UserMessage(content=user_message),
        ],
    )
    return response.choices[0].message.content
```

## Step 4 – Baseline connectivity test

A minimal prompt to confirm the deployment responds.

```python
reply = chat("Say 'connection successful' and nothing else.")
print(reply)
```

## Step 5 – Feasibility prompt

A representative prompt close to the customer scenario to validate model quality.

```python
reply = chat(
    user_message="Give me a one-line summary of why feasibility prototypes matter for engineering teams."
)
print(reply)
```
