# Prompt Test Notebook

Runs a simple chat completion against the deployed Azure OpenAI model.

```python
import os
from openai import AzureOpenAI

endpoint = os.getenv("AZURE_OPENAI_ENDPOINT")
api_key = os.getenv("AZURE_OPENAI_API_KEY")
deployment = os.getenv("AZURE_OPENAI_DEPLOYMENT", "chat")
api_version = os.getenv("AZURE_OPENAI_API_VERSION", "2024-10-21")

if not endpoint or not api_key:
    raise ValueError("Set AZURE_OPENAI_ENDPOINT and AZURE_OPENAI_API_KEY first")

client = AzureOpenAI(azure_endpoint=endpoint, api_key=api_key, api_version=api_version)
response = client.chat.completions.create(
    model=deployment,
    messages=[
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": "Give me a one-line summary of why feasibility prototypes matter."}
    ]
)
print(response.choices[0].message.content)
```
