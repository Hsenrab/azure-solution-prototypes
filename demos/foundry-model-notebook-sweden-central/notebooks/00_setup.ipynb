## Research basis

> **Why `azure-ai-inference` instead of `openai`?**
> This demo targets the **new Azure AI Foundry** (`kind: AIServices`) resource, which exposes a unified
> multi-model endpoint (`https://<name>.services.ai.azure.com/`). The `azure-ai-inference` SDK is the
> Microsoft-recommended client for this endpoint and works with any model in the AI Foundry catalog,
> not just OpenAI models. The classic `openai.AzureOpenAI` SDK still works against AIServices endpoints
> but is scoped to the OpenAI-compatible path only.
>
> **Model choice: `gpt-4.1-mini` (2025-04-14)**
> `gpt-4o-mini` was deprecated **July 2025** and retires **March 2026**.
> `gpt-4.1-mini` is its immediate replacement and is available in Sweden Central via GlobalStandard tier.
