# Step 2: Configure

Verify the deployment and check that Content Understanding endpoint and model deployment are accessible.

## Load Configuration

```python
import os
from pathlib import Path
from dotenv import load_dotenv
from azure.identity import DefaultAzureCredential
import requests

# Load .env
demo_root = Path("..").resolve()
env_file = demo_root / "env" / ".env"

print(f"Loading config from {env_file}")
load_dotenv(dotenv_path=str(env_file))

# Required env vars
required_vars = [
    "AZURE_RESOURCE_GROUP",
    "AZURE_AI_ENDPOINT",
    "CU_MODEL_DEPLOYMENT_NAME",
    "AZURE_AUTH_MODE",
]

missing = [v for v in required_vars if not os.getenv(v)]
if missing:
    raise RuntimeError(f"Missing env vars: {missing}")

print("✓ All required env vars present")
print(f"  Resource Group: {os.getenv('AZURE_RESOURCE_GROUP')}")
print(f"  AI Endpoint: {os.getenv('AZURE_AI_ENDPOINT')}")
print(f"  Deployment: {os.getenv('CU_MODEL_DEPLOYMENT_NAME')}")
print(f"  Auth Mode: {os.getenv('AZURE_AUTH_MODE')}")
```

## Check Endpoint Connectivity

```python
# Test endpoint connectivity
endpoint = os.getenv("AZURE_AI_ENDPOINT")
print(f"Testing endpoint connectivity to {endpoint}...")

try:
    response = requests.get(endpoint, timeout=5)
    print(f"✓ Endpoint reachable (HTTP {response.status_code})")
except Exception as e:
    print(f"✗ Error reaching endpoint: {e}")
    raise
```

## Verify Authentication

```python
# Get auth token
print("Testing Azure authentication...")
credential = DefaultAzureCredential()

try:
    token = credential.get_token("https://cognitiveservices.azure.com/.default")
    print(f"✓ Authentication successful")
    print(f"  Token expiry: {token.expires_on}")
except Exception as e:
    print(f"✗ Authentication failed: {e}")
    raise
```
